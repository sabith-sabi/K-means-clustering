```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

# Drop the existing 'Age' column
data = data.drop(columns='Age')

# Create a new column with random values ranging from 1 to 10
data['New_Age_Column'] = np.random.randint(1, 11, size=len(data))

# Print the modified DataFrame to see the result
data
```

```python
X = data[['New_Age_Column', 'Purchase']]

# Define a range of cluster numbers to try
max_clusters = 10  # You can change this number based on your preference
wcss = []

# Calculate the within-cluster sum of squares (WCSS) for each number of clusters
for n_clusters in range(1, max_clusters+1):
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # inertia_ stores the WCSS value

# Plot the elbow plot
plt.plot(range(1, max_clusters+1), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Plot')
plt.xticks(np.arange(1, max_clusters+1))
plt.grid(True)
plt.show()
```

elbow point is 3, choose k=3

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(X)
data['cluster'] = y_predicted
data.head()
```

```python
# Define the number of clusters (k)
n_clusters = 3

# Create a KMeans object and perform clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)

# Get the cluster assignments
cluster_assignments = kmeans.predict(X)

# Visualize the clusters
plt.scatter(X['New_Age_Column'], X['Purchase'], c=cluster_assignments, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, linewidths=2, c='black')
plt.xlabel('New_Age_Column')
plt.ylabel('Purchase')
plt.title(f'K-means Clustering with k={n_clusters}')
plt.show()
```

In this code, we use 'Age' as the x-axis and 'Purchase' as the y-axis for the scatter plot. The data points are colored based on their cluster assignments obtained from K-means. Additionally, we plot the cluster centers using 'x' markers to show the centroid of each cluster.
